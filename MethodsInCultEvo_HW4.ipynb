```python
# This function takes the source ids from the different codders sheets, and then creates a unique set that contains all coded ressources by an ID
# Expected format of the input is the list of the pages of the sheets, where each sheet is the sheet WiFa,WiMo,HuFa,HuMo of each coder
def IC_Sources_Set(frames):
  B = set()
  for frame in frames:
    # Clean 'Source_ID' values: convert to string, strip whitespace, normalize, and filter out 'nan' strings
    #Caution: This is what a Gemini proposed, as in my code without cleaning, as every member of the groups used different type of sheets documents
    # there was the error with creating different source ids..
    cleaned_source_ids = frame['Source_ID'].astype(str).str.lower().str.replace(r'\s+', ' ', regex=True).str.strip()
    cleaned_source_ids = cleaned_source_ids[cleaned_source_ids != 'nan']
    S = set(cleaned_source_ids.values)
    B = B.union(S)
  # Only remove 'column h on source_database' (after cleaning) if it exists in the set
  if 'column h on source_database' in B:
    B.remove('column h on source_database')
  return B
# print(IC_Sources_Set([df_coderM_WiFa]))
```

```python

# This function takes an individual codder sheet and the mutual of all codders sources list,
# And creates a new sheet that would containt all the sources and add the values None If the ressource was not codded by a codder
#sources_list is expected to be a set of sources as in an output from IC_Sources_Set

def IC_IndSheet(frame,sources_list):
  # Create a copy to avoid SettingWithCopyWarning
  frame_copy = frame.copy()
  # Clean 'Source_ID' column to ensure uniqueness and consistency
  frame_copy['Source_ID'] = frame_copy['Source_ID'].astype(str).str.lower().str.replace(r'\s+', ' ', regex=True).str.strip()
  # Filter out rows where 'Source_ID' became 'nan' string (from actual NaNs)
  frame_cleaned = frame_copy[frame_copy['Source_ID'] != 'nan']

  # Now drop duplicates based on the cleaned 'Source_ID' and set it as index
  frame_unique_sources = frame_cleaned.drop_duplicates(subset=['Source_ID']).set_index('Source_ID')

  result = (
    frame_unique_sources
      .reindex(sources_list)
      .reset_index()
)
  return result
```

```python
#This function take the frames from the different coders, and append them one by one


def IC_Sheet(frames,variable_name):
  sources_list = IC_Sources_Set(frames)
  new_frames=[]
  for frame in frames:
    new_frames.append(IC_IndSheet(frame,sources_list))
  new_frames = [
    df.set_index('Source_ID').rename(columns={variable_name: f"Coder_{i+1}"})
    for i, df in enumerate(new_frames)
]

  merged = pd.concat(new_frames, axis=1).reset_index()
  return merged

sources_list = IC_Sources_Set([df_coderM_WiFa,df_coderH_WiFa,df_coderD_HuFa,df_coderO_HuFa])
sources_list
```

```python
import numpy as np

def transform_value(x):
    # 1. Handle NaN (covers np.nan and float nan)
    if isinstance(x, float) and np.isnan(x):
        return np.nan

    # 2. Handle numbers (int, float but not nan)
    if isinstance(x, (int, float)):
        return float(x)

    # 3. Handle strings
    if isinstance(x, str):
        x = x.strip()

        # Special replacement
        if x == "Other [specify in evidence_quotes ]":
            return "Other"

        # Split comma-separated values
        if "," in x:
            return {item.strip() for item in x.split(",") if item.strip()}

        return {x}

    # Fallback (just return as is)
    return x


def transform_array(arr):
    vectorized = np.vectorize(transform_value, otypes=[object])
    return vectorized(arr).T
```

## This is the calculation of the ICA metric

```python
import numpy as np
from collections import Counter
import math

def krippendorff_alpha_nominal(data):
    """
    data: 2D numpy array (coders x items)
          entries can be any hashable category or np.nan
    """

    data = np.array(data, dtype=object)

    # Get all valid categories
    categories = set()
    for val in data.flatten():
        if val is not None and not (isinstance(val, float) and np.isnan(val)):
            categories.add(val)

    categories = list(categories)

    # Coincidence matrix
    coincidence = {c: Counter() for c in categories}

    for col in range(data.shape[1]):
        values = [v for v in data[:, col]
                  if v is not None and not (isinstance(v, float) and np.isnan(v))]

        n = len(values)
        if n < 2:
            continue

        for i in range(n):
            for j in range(n):
                if i != j:
                    coincidence[values[i]][values[j]] += 1

    # Observed disagreement Do
    Do = 0
    total_pairs = 0

    for c in categories:
        for k in categories:
            if c != k:
                Do += coincidence[c][k]
            total_pairs += coincidence[c][k]

    if total_pairs == 0:
        return np.nan

    Do = Do / total_pairs

    # Expected disagreement De
    marginals = Counter()
    for c in categories:
        marginals[c] = sum(coincidence[c].values())

    total = sum(marginals.values())

    De = 0
    for c in categories:
        for k in categories:
            if c != k:
                De += marginals[c] * marginals[k]

    De = De / (total * (total - 1)) if total > 1 else 0

    if De == 0:
        return 1.0

    return 1 - (Do / De)
```

```python
def normalize_multilabel(data):
    out = np.empty_like(data, dtype=object)

    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            val = data[i, j]

            if isinstance(val, set):
                out[i, j] = tuple(sorted(val))  # hashable
            else:
                out[i, j] = val

    return out

krippendorff_alpha_nominal(normalize_multilabel(IC_sheet_WiFa_avoided_interactions))
```
