```python
from numpy._core.numeric import full
base_path = '/content/kinbank/kinbank/raw'
# This code parse over all the directories and retrieve all the language filenimes over the all
all_files = []

for root, dirs, files in os.walk(base_path):
    for file in files:
        full_path = os.path.join(root, file)
        all_files.append(full_path)
del all_files[0]

all_files[1]


```

```python
#This code takes as an input the
def family_language_pair(file_name):
  return file_name.split("/")[-2:]
family_language_pair(all_files[1])
```

```python
# This code takes as an input the file name and return the name of the language,
# by simply deleting the part
def lang_name(file_name):
  return file_name.rsplit("_", 1)[0]
lang_name(family_language_pair(all_files[1])[1])
```

```python
from numpy import nan
#This code takes as an input the file name and return the richness of kin terminology,
# The number of the words in the file, because of the file structure is the number of rows
import pandas as pd
def kin_richness(file_name):
  df = pd.read_csv(file_name)
  data = list(df['word'])
  richness = 0
  for word in data:
    if type(word)==str:
        richness+=1
  return richness


df = pd.read_csv('/content/kinbank/kinbank/raw/Nuclear Trans New Guinea/Barai_nucl1630.csv')
list(df['word'])
```

```python
# This code gives our final dataset in the form:
# Language family - Language - Richness of the Kin terminology

Data =[]
for language in all_files:
  kin_rich = kin_richness(language)
  language_family = family_language_pair(language)
  language_name = lang_name(language_family[1])
  Data.append([language_family[0], language_name, kin_rich])
Data
```

```python
DataFrame = pd.DataFrame(Data, columns = ['Language family', 'Language', 'Richness of the Kin terminology'])
DataFrame
```
